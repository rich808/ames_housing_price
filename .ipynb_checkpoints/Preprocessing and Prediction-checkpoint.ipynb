```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LassoCV
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
#import clean train and test csv
test_id = pd.read_csv('./datasets/test.csv')
train = pd.read_csv('./datasets/train_clean_3_features_40.csv')
test = pd.read_csv('./datasets/test_clean_3_features_39.csv')
```

```python
#check the data for train
train.head()
```

```python
#check the data for test
test.head()
```

```python
#define function to get the list of columns name for dummies. Dtypes will be object

def all_cate(data):
    dummy = []
    for name in data.columns:
        if data[name].dtypes == 'object':
            dummy.append(name)
        else:
            pass
    return dummy
```

```python
#use testing set to get all the column names
dummy_test = all_cate(test)
```

```python
#combine the two list before get dummie to avoid different features of train and test.
#add a indentifier column to filter back afterwards. 

train['comb'] = 'train'
test['comb'] = 'test'
```

```python
#stack the 2 features into 1 big features. Default is stacking the rows.
dum = pd.concat([train, test])
```

```python
#dummies both all categorical values for train and test using the big list

dum_new = pd.get_dummies(dum, columns = dummy_test, drop_first = True)
```

```python
#split them back into test and train using the identifier column

train_d = dum_new[dum_new['comb'] == 'train']
test_d = dum_new[dum_new['comb'] == 'test']
```

```python
#drop identifier column. Dont need it anymore

train_d.drop(columns = ['comb'], inplace = True)
test_d.drop(columns = ['comb','saleprice'], inplace = True)
```

```python
#check the shap to make sure train has 1 more column than test due to 'saleprice'

train_d.shape
```

```python
test_d.shape
```

```python
#looked through the graph from EDA section for skewed charts. 

#log transformation on all the following
train_d['lot_area'] = np.log(train_d['lot_area'])
```

```python
train_d['1st_flr_sf'] = np.log(train_d['1st_flr_sf'])
```

```python
train_d['gr_liv_area'] = np.log(train_d['gr_liv_area'])
```

```python
train_d['lot_frontage'] = np.log(train_d['lot_frontage'])
```

```python
test_d['lot_area'] = np.log(test_d['lot_area'])
```

```python
test_d['1st_flr_sf'] = np.log(test_d['1st_flr_sf'])
```

```python
test_d['gr_liv_area'] = np.log(test_d['gr_liv_area'])
```

```python
test_d['lot_frontage'] = np.log(test_d['lot_frontage'])
```

```python
#define X and y for regression. Linear, Ridge, LassCV will use this

X = train_d.drop(columns = ['saleprice'])
y = np.log(train_d['saleprice'])
```

```python
#instatiate the linear regression

lr = LinearRegression()
```

```python
#train test split. With enough data, default to 75/25

X_train, X_test, y_train, y_test =train_test_split(X, y, random_state = 42)
```

```python
#score on the train set

cross_val_score(lr, X_train, y_train, cv = 5).mean()
```

```python
#score for test set

cross_val_score(lr, X_test, y_test, cv = 5).mean()
```

### Linear Regression

```python
#the scores is pretty decent. 
#to submit for kaggle, want to fit the entire train for the test set
#fit the model

lr.fit(X, y)
```

```python
#use test to predict
price_pred = lr.predict(test_d)
```

```python
price_pred
```

```python
#turn it back to regular scale. 

change = np.exp(price_pred)
change
```

### Ridge Regression

```python
#Using Regularization of Ridge
#1 Scale the data

scale = StandardScaler()
s_train = scale.fit_transform(X_train)
sc_test = scale.transform(X_test)
s_test  = scale.transform(test_d)
```

```python
#instantiate and fit with Ridge
ridge = Ridge(alpha = 1)
ridge.fit(s_train, y_train)
```

```python
ridge.score(s_train, y_train)
```

```python
ridge.score(sc_test, y_test)
```

```python
#the model was refitting with the entire training set to submit for kaggle.

r_pred = ridge.predict(s_test)
```

```python
r_pred
```

```python
#turn it back to normal scale from the log scale

r_pred_norm = np.exp(r_pred)
r_pred_norm
```

### LassoCV Regression

```python
#LASSO regularization

las = LassoCV(cv = 5)

#fit lasso

las.fit(s_train, y_train)
```

```python
las.score(s_train, y_train)
```

```python
las.score(sc_test, y_test)
```

```python
#the model was refitting with the entire train set kaggle

l_pred = las.predict(s_test)
```

```python
l_pred
```

```python
#check it back to normal scale from its log

l_pred_norm = np.exp(l_pred)
```

```python
l_pred_norm
```

```python
#clean the column names to lower case without spaces
def clean_col(data):
    name = {}
    for column in data.columns:
        name[column] = column.lower().replace(' ', '_')
    
    return data.rename(columns = name, inplace = True)
```

```python
#clean column for train
clean_col(train)
```

```python
#clean column for test
clean_col(test)
```

```python
#with the given sold year and garage built year, add age column for both. 
train['age_house'] = train['yr_sold'] - train['year_built']
train['age_garg'] = train['yr_sold'] - train['garage_yr_blt']
train['age_rem'] = train['yr_sold'] - train['year_remod/add']

#check 
train.head()
```

```python
#do the same for test set
test['age_house'] = test['yr_sold'] - test['year_built']
test['age_garg'] = test['yr_sold'] - test['garage_yr_blt']
test['age_rem'] = test['yr_sold'] - test['year_remod/add']
```

```python
#functiont to plot subplots
def b_plot(data):
    rows = int(np.ceil(len(data)/2)) 
    fig, ax = plt.subplots(rows, 2, figsize = (10,20))
    ax = ax.ravel()
    for i, e in enumerate(data):
        sns.boxplot(x = train[e], y =train['saleprice'], ax = ax[i])
        ax[i].set_title(e)
        ax[i].set_xlabel(e)
        
    plt.tight_layout()
```

```python
#divide box plot execution into sections. 

# part 1
b_plot(keep_qn[:12])
```

```python
#part 2

b_plot(keep_qn[13:27]) 
```

```python
#part 3
b_plot(keep_qn[27:33]) 
```

```python
#Do the same for test set but remove saleprice

for_test = ['age_house', 'age_garg', 'age_rem', 'lot_area', 'wood_deck_sf', 'open_porch_sf', 
        'lot_frontage', 'bsmtfin_sf_1', 'fireplaces', 'totrms_abvgrd', 
       'mas_vnr_area', 'full_bath', '1st_flr_sf', 'total_bsmt_sf',
       'garage_cars', 'garage_area', 'gr_liv_area', 'overall_qual'] + keep 
```

```python
#turn the list into features dataframe
features_test = test[for_test]
```

```python
#check for test data type
features_test.dtypes
```

```python
#check for null on the test set.
features_test.isnull().sum()
```

```python
#replace lot_frontage with the means corresponding to the neighborhood

#1 get a dictionary of mean
front = {}
for neigh in features_test['neighborhood'].unique():
    front[neigh] = features_test[features_test['neighborhood'] == neigh]['lot_frontage'].mean()

#2 replace each NAN with its mean
features_test['lot_frontage'][features_test['lot_frontage'].isnull()] = features_test['neighborhood'][features_test['lot_frontage'].isnull()].map(front)
```

```python
#fill garage_age and mas_vnr_type with 0 since is either 0 or have some value to it. 

features_test.loc[features_test['age_garg'].isnull(), 'age_garg'] = 0
```

```python
features_test.loc[features_test['mas_vnr_type'].isnull(), 'mas_vnr_type'] = 0
```

```python
features_test.loc[features_test['mas_vnr_area'].isnull(), 'mas_vnr_area'] = 0
```

```python
#drop 'GrnHill' and 'Landmrk' for features since only have it in train data and test set doesn't have it
features = features[(features['neighborhood'] != 'GrnHill') & (features['neighborhood'] != 'Landmrk')]
```
